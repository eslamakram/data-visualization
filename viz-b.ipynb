```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # Set default plotter to be seaborn
```

```python
data = sns.load_dataset('mpg')
data.head(30)
```

## The avarage for horsepower of car based on thier origin countries


```python
filter_data = data.groupby("origin", as_index=False).mean()
sns.barplot(x="origin",y="horsepower",data=filter_data)
```

## The avarage for horsepower of car based on its model year


```python
filter_data = data.groupby("model_year", as_index=False).mean()
sns.barplot(x="model_year",y="horsepower",data=filter_data)
```

## The avarage for horsepower of car based on its mpg

```python
filter_data = data.groupby("mpg", as_index=False).mean()
sns.barplot(x="mpg",y="horsepower",data=filter_data)
```

### summary
* The most horsepower of car are The car which manifactures in USA
* Based on the model year of the car; 70's have high horsepower engine then the other years
* Japan and europe cars has the same horsepower
